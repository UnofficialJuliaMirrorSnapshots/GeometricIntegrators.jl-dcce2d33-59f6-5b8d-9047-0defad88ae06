# Test Bernstein Polynomials

```julia
using GeometricIntegrators
```

```julia
using Gadfly
```

Number of nodes:

```julia
s = 5
```

Number of points to plot:

```julia
np = 101
```

Obtain Gauss-Legendre nodes:

```julia
c = collect(linspace(0, 1, s))
```

Define Lagrange polynomial:

```julia
ber = BernsteinBasis(c)
```

Compute plot range:

```julia
xp = collect(linspace(0, 1, np));
```

Plot all basis polynomials:

```julia
layers = []

for i in 1:nbasis(ber)
    yp = zeros(np)
    for j in eachindex(xp,yp)
        yp[j] = evaluate(ber, i, xp[j])
    end
    push!(layers, layer(x=xp, y=yp, Geom.line))
end

plot(layers...)
```

```julia
layers = []

for i in 1:nbasis(ber)
    yp = zeros(np)
    for j in eachindex(xp,yp)
        yp[j] = derivative(ber, i, xp[j])
    end
    push!(layers, layer(x=xp, y=yp, Geom.line))
end

plot(layers...)
```
