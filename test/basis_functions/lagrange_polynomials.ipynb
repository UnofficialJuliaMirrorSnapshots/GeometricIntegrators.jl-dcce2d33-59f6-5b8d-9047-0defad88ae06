# Test Lagrange Polynomials

```julia
using GeometricIntegrators
```

```julia
using Gadfly
```

Number of nodes:

```julia
s = 3
```

Number of points to plot:

```julia
np = 101
```

Obtain Gauss-Legendre nodes:

```julia
gl = GaussLegendreQuadrature(s)
c  = nodes(gl)
```

Define Lagrange polynomial:

```julia
lag = LagrangeBasis(c)
```

Compute plot range:

```julia
xp = collect(linspace(0., 1., np));
```

Plot all basis polynomials:

```julia
layers = []

for i in 1:nbasis(lag)
    yp = zeros(np)
    for j in eachindex(xp,yp)
        yp[j] = evaluate(lag, i, xp[j])
    end
    push!(layers, layer(x=xp, y=yp, Geom.line))
end

plot(layers...)
```

```julia
layers = []

for i in 1:nbasis(lag)
    yp = zeros(np)
    for j in eachindex(xp,yp)
        yp[j] = derivative(lag, i, xp[j])
    end
    push!(layers, layer(x=xp, y=yp, Geom.line))
end

plot(layers...)
```
